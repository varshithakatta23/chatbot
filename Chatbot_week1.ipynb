# Collect and clean the Dataset

- Normalization
- Removing Special Characters
- Stopword Removal
- Stemming or Lemmatization

- Removing Duplicate Records

**Tools and Libraries**
- NLTK (Python)
- spaCy (Python)
- Regular Expressions (regex)
- JSON libraries

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import json

nltk.download('punkt')
nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()

# Load JSON data
with open('intents.json') as f:
    data = json.load(f)

# Normalize text
for item in data:
    # Check if 'text' key exists before accessing it
    if 'text' in item:
        item['text'] = word_tokenize(item['text'])
        item['text'] = [lemmatizer.lemmatize(word) for word in item['text']]
        item['text'] = ' '.join(item['text'])
    # You can handle cases where 'text' is missing,
    # for example, by printing a message or skipping the item:
    else:
        print(f"Warning: 'text' key not found in item: {item}")
```

*Tools and Libraries:

- Tokenization: nltk.word_tokenize() or spaCy.tokenizer

- Stopword removal: nltk.corpus.stopwords or spaCy.lang.en.stop_words

- Stemming/Lemmatization: nltk.stem or spaCy.lemmatizer

- Frequency analysis: collections.Counter

- Visualization: matplotlib, seaborn, or plotly

- Corpus analysis: gensim or scikit-learn

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from collections import Counter
import json

# Download the stopwords resource if not already downloaded
nltk.download('stopwords')
# Download the 'punkt_tab' resource
nltk.download('punkt_tab')

# Load JSON data
with open('intents.json') as f:
    data = json.load(f)

# Tokenize text fields
tokenized_texts = []
for obj in data:
    # Check if 'patterns' key exists (original structure)
    if 'patterns' in obj:
        # Access and process 'patterns' if available
        for pattern in obj['patterns']:
            tokens = word_tokenize(pattern)
            tokenized_texts.append(tokens)
    # If 'patterns' is not there, it might be the modified structure
    else:
        # Assuming 'text' is the key holding the processed text
        tokens = word_tokenize(obj['text'])
        tokenized_texts.append(tokens)

# Remove stopwords
stop_words = set(stopwords.words('english'))
tokenized_texts = [[t for t in tokens_iter if t not in stop_words] for tokens_iter in tokenized_texts]

# Calculate frequency of each token
freq_dict = Counter(token for tokens in tokenized_texts for token in tokens)

# Visualize token frequency
import matplotlib.pyplot as plt
plt.bar(freq_dict.keys(), freq_dict.values())
plt.xlabel('Token')
plt.ylabel('Frequency')
plt.show()
```

# Split data into training , validation , and test sets

 Step 1: Load and Preprocess the JSON Data
- Load the JSON file into a Pandas DataFrame using pandas.read_json().
- Ensure the DataFrame has a consistent structure, with each row representing a text sample and columns for text data and labels (if applicable).
- Perform any necessary preprocessing steps, such as:
  - Tokenization (e.g., using NLTK or spaCy)
  - Stopword removal
  - Stemming or Lemmatization
  - Encoding (e.g., converting categorical labels to numerical values)

Step 2: Split the Data into Training, Validation, and Testing Sets
- Use the train_test_split function from scikit-learnâ€™s model_selection module to split your preprocessed DataFrame into three sets:
  - Training set (e.g., 70% of the data)
  - Validation set (e.g., 15% of the data)
  - Testing set (e.g., 15% of the data)
- Specify the test_size parameter to control the proportion of samples in each set. For example, test_size=0.3 would allocate 30% of the data to the testing set.
- Set the random_state parameter to ensure reproducibility and consistency across splits.

Step 3: Save the Split Data

```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from sklearn.model_selection import train_test_split

# Load raw JSON file
df = pd.read_json('intents.json')

# Create a 'text' column by combining patterns
df['text'] = df['patterns'].apply(lambda patterns: ' '.join(patterns))

# Preprocess text data
df['text'] = df['text'].apply(word_tokenize)
stop_words = set(nltk.corpus.stopwords.words('english'))
df['text'] = df['text'].apply(lambda x: [word for word in x if word not in stop_words])

# Split data into training, validation, and testing sets
train_text, val_test_text, train_labels, val_test_labels = train_test_split(df['text'], df['tag'], test_size=0.2, random_state=42) # Assuming 'tag' is the label column

# Write split data to JSON files
with open('training_set.json', 'w') as f:
    json.dump({'text': train_text.tolist(), 'labels': train_labels.tolist()}, f)

with open('validation_set.json', 'w') as f:
    json.dump({'text': val_test_text[:len(val_test_labels)//2].tolist(), 'labels': val_test_labels[:len(val_test_labels)//2].tolist()}, f)

with open('testing_set.json', 'w') as f:
    json.dump({'text': val_test_text[len(val_test_labels)//2:].tolist(), 'labels': val_test_labels[len(val_test_labels)//2:].tolist()}, f)
```
